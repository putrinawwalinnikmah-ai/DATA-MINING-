```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_excel("Jumlah_Sekolah_Akses_Internet_2024_reformatted.xlsx")

# Tambahkan kolom persentase akses internet
df["Persentase_Akses"] = (df["Jumlah_Internet"] / df["Jumlah_Sekolah"]) * 100

df.head()
```

Persentase Akses Internet Data 20 Pertama

```python
# nomor data
x = range(1,21)
y = df["Persentase_Akses"].head(20)

plt.plot(x, y, 'go-')
plt.xlabel('Nomor Data')
plt.ylabel('Persentase Akses Internet (%)')
plt.title('Grafik Persentase Akses Internet (20 Data Pertama)')
plt.show()
```

Bar Chart Total Sekolah per Provinsi

```python
# Mengelompokkan data berdasarkan Provinsi
total_provinsi = df.groupby("Provinsi")["Jumlah_Sekolah"].sum()

plt.figure(figsize=(13,6))
total_provinsi.sort_values().plot(kind="bar")

plt.grid()
plt.title("Total Sekolah dengan Akses Internet per Provinsi Tahun 2024")
plt.xlabel("Provinsi")
plt.ylabel("Jumlah Sekolah")
plt.xticks(rotation=90)

plt.show()
```

PIE CHART

```python
# PIE CHART PERSENTASE PER JENJANG
total_jenjang = df.groupby('Pendidikan')['Jumlah_Sekolah'].sum()

plt.pie(total_jenjang, labels=total_jenjang.index, autopct='%1.1f%%')
plt.title("Persentase Sekolah Akses Internet per Jenjang")

plt.show()
```

```python
# Mengelompokkan berdasarkan Jenjang
total_jenjang = df.groupby("Pendidikan")["Jumlah_Sekolah"].sum()

x = total_jenjang.index
y = total_jenjang.values

plt.plot(x, y, 'g*-')
plt.xlabel('Jenjang')
plt.ylabel('Jumlah Sekolah')
plt.title('Diagram Garis Total Sekolah Akses Internet per Jenjang 2024')

plt.show()
```

STACKED BAR CHART PROVINSI DAN JENJANG

```python
# STACKED BAR CHART PROVINSI DAN JENJANG

pivot_data = df.pivot_table(index="Provinsi",
                              columns="Pendidikan",
                              values="Jumlah_Sekolah",
                              aggfunc="sum")

pivot_data.plot(kind="bar", stacked=True, figsize=(14,6))

plt.title("Distribusi Sekolah Akses Internet per Provinsi dan Jenjang")
plt.xlabel("Provinsi")
plt.ylabel("Jumlah Sekolah")
plt.xticks(rotation=90)

plt.show()
```

Exploded Pie Plot

```python
# Mengelompokkan data berdasarkan Jenjang
total_jenjang = df.groupby("Pendidikan")["Jumlah_Sekolah"].sum()

data_pie = total_jenjang.values
kategori = total_jenjang.index

warna = ['pink', 'cyan', 'yellowgreen', 'skyblue', 'purple']
explode_var = [0.1 if i == 2 else 0 for i in range(len(data_pie))]

plt.pie(data_pie,
        labels=kategori,
        colors=warna[:len(data_pie)],
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)

plt.legend()
plt.title('Exploded Pie Plot Sekolah Akses Internet per Jenjang 2024')

plt.show()
```
